### Part 1 Analyze the log file without option file
Option File helps to break the users into CCU and NNU

1.1 Import Support Package

```python
import pandas as pd
```

```python
file1 = open('matlab_servicelog_2019a.log', 'r')
Lines = file1.readlines()
Result = []
time = "Unavailable\n"
```

1.2 Text Encoding

```python
for line in Lines:
    W=line.split(" ")[0:5]
    if "TIMESTAMP" in line:
        time=W[-1]
    if len(W)>2:
        if W[2]=='OUT:' or W[2]=="IN:" or W[2]=="DENIED:":
            Result.append([time[0:-1],W[2],W[3],W[4]])
```

```python
df = pd.DataFrame(Result,columns=['Date','Status','Toolbox','User'])
print (df)
```

1.3 Save to Excel

```python
df.to_excel (r'report1.xlsx', index = False, header=True)
```
